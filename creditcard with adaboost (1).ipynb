let's check outliers on the basis of the skewness of the column.

```python
def outliersiqr(d,j):
    q1 = d[j].quantile(0.25)
    q3 = d[j].quantile(0.75)
    iqr = q3 - q1
    low = q1 - (1.5*iqr)
    hi = q3 + (1.5*iqr)
    l1 = d[(d[j]<= low) | (d[j]>= hi)]
    v1 = d[(d[j]>= low) & (d[j]<= hi)]
    return l1,v1
```

```python
def outliesmsd(d,j):
    m = d[j].mean()
    sd = d[j].std()
    low = m - (3*sd)
    hi = m + (3*sd)
    ot= d[(d[j]<=low) | (d[j]>=hi)]
    vld =  d[(d[j]>=low) & (d[j]<=hi)] 
    return ot, vld
```
